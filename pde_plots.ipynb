```python
import numpy as np
import pandas as pd  # assuming df_euler is a pandas DataFrame

# ==============================
# Shock-capturing & Accuracy Metrics
# ==============================

def shock_location(x, var, window=None):
    """Estimate shock position = x at maximum |gradient|."""
    if window is not None:
        mask = (x >= window[0]) & (x <= window[1])
        x_sel, var_sel = x[mask], var[mask]
    else:
        x_sel, var_sel = x, var
    if len(x_sel) < 2:
        return np.nan
    grad = np.abs(np.gradient(var_sel, x_sel))
    return x_sel[np.argmax(grad)]


def shock_thickness(x, var, window=None, low=0.1, high=0.9):
    """10%–90% transition width of the discontinuity."""
    if window is not None:
        mask = (x >= window[0]) & (x <= window[1])
        x_sel, var_sel = x[mask], var[mask]
    else:
        x_sel, var_sel = x, var

    v_min, v_max = np.min(var_sel), np.max(var_sel)
    jump = v_max - v_min
    if jump < 1e-8:
        return 0.0

    vnorm = (var_sel - v_min) / jump
    mask_trans = (vnorm >= low) & (vnorm <= high)

    if np.sum(mask_trans) < 3:
        return 0.0

    x_trans = x_sel[mask_trans]
    return x_trans[-1] - x_trans[0]


def post_shock_oscillation(num, exact, post_region):
    """Maximum absolute deviation in the post-shock region."""
    mask = (x >= post_region[0]) & (x <= post_region[1])
    if np.sum(mask) == 0:
        return np.nan
    return np.max(np.abs(num[mask] - exact[mask]))


def total_variation(var, region=None):
    """Total variation in the specified region (or full domain)."""
    if region is not None:
        mask = (x >= region[0]) & (x <= region[1])
        var_sel = var[mask]
    else:
        var_sel = var
    if len(var_sel) < 2:
        return 0.0
    return np.sum(np.abs(np.diff(var_sel)))


def l1_error(num, exact):
    """Mean absolute error (L1 norm)."""
    return np.mean(np.abs(num - exact))


def linf_error(num, exact):
    """Maximum absolute error (L∞ norm)."""
    return np.max(np.abs(num - exact))


# ==============================
# Load data from DataFrame
# ==============================

x = df_euler["x"].values

rho_exact = df_euler["rho_exact"].values
u_exact   = df_euler["u_exact"].values
P_exact   = df_euler["P_exact"].values

rho_kbr   = df_euler["rho_present"].values
rho_god   = df_euler["rho_fdm"].values
rho_muscl = df_euler["rho_muscl"].values

u_kbr     = df_euler["u_present"].values
u_god     = df_euler["u_fdm"].values
u_muscl   = df_euler["u_muscl"].values

P_kbr     = df_euler["P_present"].values
P_god     = df_euler["P_fdm"].values
P_muscl   = df_euler["P_muscl"].values


# ==============================
# Compute metrics for one variable
# ==============================

def compute_metrics(var_name, var_exact, schemes_dict, shock_window, post_region):
    """
    Compute metrics for all schemes.
    Returns list of dicts.
    """
    shock_exact = shock_location(x, var_exact, shock_window)

    results = []
    for scheme_name, var_num in schemes_dict.items():
        row = {
            "Scheme": scheme_name,
            "L1 Err":     l1_error(var_num, var_exact),
            "L∞ Err":      linf_error(var_num, var_exact),
            "Thickness":   shock_thickness(x, var_num, shock_window),
            "Post Osc":    post_shock_oscillation(var_num, var_exact, post_region),
            "TV":          total_variation(var_num, shock_window)
        }
        results.append(row)
    return results


# ==============================
# Scheme groups
# ==============================

rho_schemes = {
    "KBR-Roe":   rho_kbr,
    "Roe":       rho_god,
    "MUSCL-TVD": rho_muscl
}

u_schemes = {
    "KBR-Roe":   u_kbr,
    "Roe":       u_god,
    "MUSCL-TVD": u_muscl
}

P_schemes = {
    "KBR-Roe":   P_kbr,
    "Roe":       P_god,
    "MUSCL-TVD": P_muscl
}


# ==============================
# Pretty-print function
# ==============================

def print_metrics_table(title, results):
    print(f"\n{title}\n")
    header = (
        f"{'Scheme':<12} "
        f"{'L1 Err':>10} "
        f"{'L∞ Err':>10} "
        f"{'Thickness':>12} "
        f"{'Post Osc':>10} "
        f"{'TV':>10}"
    )
    print(header)
    print("-" * len(header))

    for r in results:
        print(
            f"{r['Scheme']:<12} "
            f"{r['L1 Err']:>10.3e} "
            f"{r['L∞ Err']:>10.3e} "
            f"{r['Thickness']:>12.3e} "
            f"{r['Post Osc']:>10.3e} "
            f"{r['TV']:>10.3e}"
        )


# ==============================
# Define windows (unchanged as requested)
# ==============================

# First region (likely main shock / contact area)
SHOCK_WINDOW_1    = [0.62, 0.67]
POST_SHOCK_1      = [0.64, 0.72]

# Second region (likely another discontinuity / secondary feature)
SHOCK_WINDOW_2    = [0.73, 0.77]
POST_SHOCK_2      = [0.755, 0.8]


# ==============================
# Compute and display for first region
# ==============================

print("=== First Region ===")
rho_results_1 = compute_metrics("rho", rho_exact, rho_schemes, SHOCK_WINDOW_1, POST_SHOCK_1)
print_metrics_table("Density (ρ) - Region 1", rho_results_1)

# Optionally compute u and P for the same region
# u_results_1   = compute_metrics("u",   u_exact,   u_schemes,   SHOCK_WINDOW_1, POST_SHOCK_1)
# P_results_1   = compute_metrics("P",   P_exact,   P_schemes,   SHOCK_WINDOW_1, POST_SHOCK_1)
# print_metrics_table("Velocity (u) - Region 1", u_results_1)
# print_metrics_table("Pressure (P) - Region 1", P_results_1)


# ==============================
# Compute and display for second region
# ==============================

print("\n=== Second Region ===")
rho_results_2 = compute_metrics("rho", rho_exact, rho_schemes, SHOCK_WINDOW_2, POST_SHOCK_2)
print_metrics_table("Density (ρ) - Region 2", rho_results_2)

# Optionally for u and P
# u_results_2   = compute_metrics("u",   u_exact,   u_schemes,   SHOCK_WINDOW_2, POST_SHOCK_2)
# P_results_2   = compute_metrics("P",   P_exact,   P_schemes,   SHOCK_WINDOW_2, POST_SHOCK_2)
# print_metrics_table("Velocity (u) - Region 2", u_results_2)
# print_metrics_table("Pressure (P) - Region 2", P_results_2)
```
